```python
import pandas as pd
from IPython import display as ICD
import numpy as np
```

```python
games = pd.read_csv('./predictions/games.csv')
```

```python
games['WinProba'] = games['Odds'].apply(lambda x: -x/(100-x))
games
```

```python
target = 1.04
```

```python
games['HomeWinProba'] = games.apply(lambda x: x['WinProba'] if x['TeamName'] == x['Winningteam'] else target - x['WinProba'], axis = 1)
games['AwayWinProba'] = games.apply(lambda x: target - x['WinProba'] if x['TeamName'] == x['Winningteam'] else x['WinProba'], axis = 1)
games['HomeReturn'] = games['HomeWinProba'].apply(lambda x: (1-x)/x)
games['AwayReturn'] = games['AwayWinProba'].apply(lambda x: (1-x)/x)
games
```

```python
cleaned = games[(~pd.isna(games['TeamPointsActual'])) & (~pd.isna(games['HomeWinProba']))].iloc[8:]
```

```python
def returns(row, threshold):
    if(row['TeamWinProba'] - row['HomeWinProba'] >= threshold):
        if(row['TeamPointsActual'] > row['OpponentTeamPointsActual']):
            return row['HomeReturn']
        else:
            return -1
    if(row['OpponentTeamWinProba'] - row['AwayWinProba'] >= threshold):
        if(row['TeamPointsActual'] < row['OpponentTeamPointsActual']):
            return row['AwayReturn']
        else:
            return -1
    return 0
def get_returns(threshold, display):
    temp = cleaned.apply(lambda x: returns(x, threshold), axis = 1)
    temp_two = pd.concat([cleaned, temp], axis = 1)
    if(display):
        ICD.display(temp_two[temp_two[0] != 0][['GameNum', 'TeamName',
       'OpponentTeamName', 'TeamWinProba', 'OpponentTeamWinProba','HomeWinProba', 'AwayWinProba', 'TeamPointsActual',
       'OpponentTeamPointsActual',0]])
    return temp[temp != 0].mean()
    
def get_count(threshold, display):
    temp = cleaned.apply(lambda x: returns(x, threshold), axis = 1)
    temp_two = pd.concat([cleaned, temp], axis = 1)
    return len(temp[temp != 0])
```

```python
results = pd.DataFrame()
results['Threshold'] = np.arange(0,0.25,0.01)
results['Returns'] = results['Threshold'].apply(lambda x: get_returns(x, False))
results['Count'] = results['Threshold'].apply(lambda x: get_count(x, False))


results
```
